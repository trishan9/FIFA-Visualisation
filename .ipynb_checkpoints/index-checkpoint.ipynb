# Fifa Data Visualisation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fifa = pd.read_csv("./fifa_eda.csv")
fifa
```

## Pie Chart Example

```python
plt.figure(figsize=(5, 5))
# Creating a pie chart for the top 5 nationalities in the FIFA dataset
plt.pie(fifa['Nationality'].value_counts().head(5), labels=fifa['Nationality'].value_counts().head(5).index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Nationalities in FIFA Dataset')  # Adding a title to the pie chart
plt.show()
```

## Histogram Example

```python
plt.figure(figsize=(10, 5))
# Plotting a histogram for the age distribution in the FIFA dataset using seaborn
sns.histplot(fifa['Age'], bins=30, kde=True, color='red')
plt.title('Histogram Example - Age Distribution in FIFA Dataset')  # Adding a title to the histogram
plt.xlabel('Age')  # Labeling the X-axis as player age
plt.ylabel('Frequency')  # Labeling the Y-axis as frequency
plt.show()
```

## Box Plot Example

```python
plt.figure(figsize=(8, 6))
# Creating a box plot for the overall rating based on the preferred foot in the FIFA dataset
sns.boxplot(x='Preferred Foot', y='Overall', data=fifa)
plt.title('Box Plot Example - Overall Rating by Preferred Foot in FIFA Dataset')  # Adding a title to the box plot
plt.xlabel('Preferred Foot')  # Labeling the X-axis as preferred foot
plt.ylabel('Overall Rating')  # Labeling the Y-axis as overall rating
plt.show()
```

## Linear Regression Example

```python
plt.figure(figsize=(10, 6))
# Creating a scatter plot with a linear regression line for the relationship between age and overall rating
sns.regplot(x='Age', y='Overall', data=fifa, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Linear Regression - Age vs Overall Rating in FIFA Dataset')  # Adding a title to the regression plot
plt.xlabel('Age')  # Labeling the X-axis as player age
plt.ylabel('Overall Rating')  # Labeling the Y-axis as overall rating
plt.show()
```

## Heatmap Example

```python
df_corr = fifa[['Overall','Value','Wage']].corr()
sns.heatmap(df_corr,annot=True)
```
